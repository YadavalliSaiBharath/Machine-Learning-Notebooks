## Using Imblearn (Random Under Sampling, Random Over Sampling) 

### Random Under Sampling

```python
id1 = pd.read_csv("Social_Network_Ads.csv")
id1 = pd.DataFrame(id1)
id1.head(3)
```

```python
id1["Purchased"].value_counts()
```

```python
xx = id1.iloc[:,:-1]
yy = id1["Purchased"]
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru1 = RandomUnderSampler()
ru_x , ru_y = ru1.fit_resample(xx,yy)
```

```python
ru_y.value_counts() 
# We Made 0 and 1 Equal
```

```python
from sklearn.model_selection import train_test_split
```

```python
xxtrain,xxtest,yytrain,yytest = train_test_split(ru_x,ru_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
L11 = LogisticRegression()
L11.fit(xxtrain,yytrain)
```

```python
L11.score(xxtest,yytest)*100
```

```python
L11.predict([[46,28000]])
```

```python
L11.predict([[19,19000]])
```

### Random Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
ro_x ,ro_y = ro.fit_resample(xx,yy)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xx_train1 , xx_test1, yy_train1 , yy_test1 = train_test_split(ro_x,ro_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lo = LogisticRegression()
Lo.fit(xx_train1,yy_train1)
```

```python
Lo.score(xx_test1,yy_test1)*100
```

```python
Lo.predict([[19,19000]])
```

```python
Lo.predict([[45,26000]])
```

```python
Lo.predict([[46,28000]])
```

# Naive Bayes (Gaussian , Bernoulli, Multinomial)

```python
nb = pd.read_csv("Social_Network_Ads.csv")
nb.head(3)
```

```python
nb["Purchased"].value_counts()
```

```python
nb.columns
```

```python
sns.kdeplot(data= nb["Age"])
plt.show()
```

```python
sns.kdeplot(data= nb["EstimatedSalary"])
plt.show()
```

```python
sns.scatterplot(x = "Age",y = "EstimatedSalary",data = nb,hue = "Purchased")
plt.show()
```

```python
x = nb.iloc[:,:-1]
y = nb["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
from mlxtend.plotting import plot_decision_regions
```

### Gaussian

```python
ga = GaussianNB()
ga.fit(xtrain,ytrain)
```

```python
ga.score(xtest,ytest)*100 , ga.score(xtrain,ytrain)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf= ga)
plt.show()
```

### Multinomial

```python
mt = MultinomialNB()
mt.fit(xtrain,ytrain)
```

```python
mt.score(xtest,ytest)*100 , mt.score(xtrain,ytrain)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf= mt)
plt.show()
```

### Bernoulli

```python
br = BernoulliNB()
br.fit(xtrain,ytrain)
```

```python
br.score(xtest,ytest)*100 , br.score(xtrain,ytrain)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf= br)
plt.show()
```
